**Note numpy slicing creates a view and not a copy and fancy and boolean indexing create a copy**

In case of view changes made in new array will reflect in original array
In case of copy changes made in new will not reflect in original array.

```python
a=np.array([1,2,3])
b=a[1:]
print(a)
print(b)
b[0]=10
print(a)
print(b)
```

**The View (Memory Efficient, but Dangerous)**

When you slice an array (a[:5]), NumPy doesn’t allocate new memory or duplicate data. It simply points to the exact same memory block of the original array.

Why it’s great: It is incredibly fast and memory efficient. You can handle gigabytes of data without running out of RAM.

The Trap: If you modify the view, you mutate the original data. This leads to silent, nightmare-inducing bugs in your data pipelines.

**The Copy (Safe, but Costs Memory)**

A copy creates a completely independent array located in a brand-new block of memory.

The Fix: If you need to manipulate a slice without touching the original source of truth, you must explicitly demand a copy.

```python
original_array = np.array([10, 20, 30, 40, 50])
safe_subset = original_array[:3].copy()
safe_subset[0] = -1
print(original_array)
print(safe_subset)
```
